# 低回收率问卷的辅助信息估计工具

本 Notebook 展示如何检查回收样本偏差，并用辅助变量提升总体指标估计稳定性。

```python
import pandas as pd
import matplotlib.pyplot as plt
from src.auxiliary_estimation import check_response_bias, predict_and_calibrate

df = pd.read_csv('../data/sample_low_response_survey.csv')
check_response_bias(df)
```

```python
estimate, df_pred, model = predict_and_calibrate(df)
naive_estimate = df[df.is_responded]['true_response'].mean()
full_data_reference = df['true_response'].mean()
print({'naive_sample_mean': naive_estimate, 'calibrated_estimate': estimate, 'full_data_reference': full_data_reference})
```

```python
bias = check_response_bias(df)
ax = bias.set_index('group')[['x_score_mean','auxiliary_score_mean']].plot(kind='bar')
ax.set_title('Response Bias Check')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../figures/response_bias_check.png', dpi=200)
plt.show()
```

## 业务解释

如果回收样本与全量样本在关键辅助变量上存在差异，应在报告中明确样本代表性风险，并通过辅助信息估计降低决策误差。
