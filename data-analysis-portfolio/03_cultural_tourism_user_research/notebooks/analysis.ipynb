```python
import pandas as pd
import matplotlib.pyplot as plt
from src.survey_analysis import crosstab_analysis, group_summary

df = pd.read_csv('../data/sample_tourism_survey.csv')
df.head()
```

```python
table, test = crosstab_analysis(df, 'game_user', 'purchase_intention')
display(table)
test
```

```python
plot_data = pd.crosstab(df['game_user'], df['purchase_intention'], normalize='index')
ax = plot_data.plot(kind='bar')
ax.set_title('Purchase Intention by Game User Group')
ax.set_ylabel('Proportion')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('../figures/purchase_intention_by_group.png', dpi=200)
plt.show()
```
