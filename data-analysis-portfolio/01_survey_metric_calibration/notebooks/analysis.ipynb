# 小样本调查数据校准与总体指标估计

本 Notebook 演示从数据读取、特征处理、模型预测到校准估计的完整流程。

```python
import pandas as pd
import matplotlib.pyplot as plt
from src.calibration_pipeline import load_data, train_predictor, evaluate

df = load_data('../data/sample_survey_metrics.csv')
df.head()
```

```python
df_pred, model = train_predictor(df)
result = evaluate(df_pred)
result
```

```python
ax = result.plot(kind='bar', x='method', y='estimate', legend=False)
ax.set_title('Estimation Comparison')
ax.set_ylabel('Estimated Metric')
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig('../figures/metric_comparison.png', dpi=200)
plt.show()
```

## 业务解释

直接预测可能因为模型偏差而产生系统性误判；校准估计通过少量真实标签修正预测偏差，更适合用于低标注成本场景下的总体指标估计。
